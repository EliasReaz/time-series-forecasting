```python
# === IMPORTS ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import optuna
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit
import shap
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")

# === 1. LOAD & CLEAN DATA ===
def load_data(filepath):
    """Load, then clean, and resample time series data to daily frequency."""
    df = pd.read_csv(filepath)
    df['Timestamp'] = pd.to_datetime(df['Timestamp'], format='%Y-%m-%dT%H:%M:%S')
    df.drop(columns=['_id'], inplace=True, errors='ignore')
    df.set_index('Timestamp', inplace=True)
    df = df.resample('D').sum()
    df = df[(df.index >= "2022-01-01") & (df.index <= "2025-06-30")]
    return df

def plot_original_time_series(df, target_col="Redemption Count"):
    fig, ax = plt.subplots(1,1, figsize=(12, 5))
    ax.scatter(df.index, df[target_col], s=8, color='steelblue', alpha=1, label=target_col)
    ax.set_xlabel("Date")
    ax.set_ylabel(target_col)
    ax.plot(df.index, df[target_col].shift(1).rolling(30).mean(), color='orange', alpha=1, label='30-Day Moving Average', linewidth=2)

    ax.set_xlabel("Date")
    ax.set_ylabel(f"30-Day Rolling Mean {target_col}")
    # Set the x-axis major locator to year and formatter to date format
    # Set x-ticks every year begin with the first date in the dataset
    xticks = pd.date_range(start=df.index.min(), end=df.index.max(), freq='Y')
    ax.set_xticks(xticks)
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.xaxis.set_major_locator(mdates.YearLocator(1))


    plt.title(f"Daily {target_col} & 30-Day Moving Average Starting From 2022-01-01", fontsize=17)
    plt.legend(loc='upper right', fontsize=10)
    plt.grid()
    plt.tight_layout()
    plt.show() 

# === 2. FEATURE ENGINEERING ===
def create_features(df, target_col, non_target_col):
    df['dayofweek'] = df.index.dayofweek
    df['lag1'] = df[target_col].shift(1)
    df['rolling7'] = df[target_col].shift(1).rolling(window=7).mean()
    df['rolling14'] = df[target_col].shift(1).rolling(window=14).mean()
    # featurename = 'non_target_col' + 'rolling7'
    df[non_target_col + '_rolling7'] = df[non_target_col].shift(1).rolling(window=7).mean()
    df['is_weekend'] = df['dayofweek'].isin([5, 6]).astype(int)
    df['is_summer'] = df.index.month.isin([5, 6, 7]).astype(int)
    df['is_summer_weekend'] = df.index.month.isin([5, 6, 7]).astype(int) * df['is_weekend']
    df['lag1'] = df[target_col].shift(1)

    dayofyear = df.index.dayofyear
    df['sine_annual'] = np.sin(2 * np.pi * dayofyear / 365)
    df['cosine_annual'] = np.cos(2 * np.pi * dayofyear / 365)
    # Model learn local volatility
    # df['rolling_max7'] = df[target_col].shift(1).rolling(window=7).max()
    # df['rolling_std7'] = df[target_col].shift(1).rolling(window=7).std()

    df.dropna(inplace=True)
    return df


# === 3. MODEL EVALUATION ===
def evaluate_model(y_true, y_pred, label="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    print(f"{label} → MAE: {mae:.2f}, MSE: {mse:.2f}, RMSE: {rmse:.2f}")
    return mae, mse, rmse


def plot_predictions(y_true, base_pred_original, index, fold):
    fig, ax = plt.subplots(1, 1, figsize=(15, 8))
    ax.plot(index, y_true, label='Actual', color='orange', marker='o')
    ax.plot(index, base_pred_original, label='Base Prediction', color='green', linestyle='--', marker='o')
    ax.set_title(f"Fold {fold} - Base Light XGBoost")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()
 
# === Add plot_with_CI() function ===    
def plot_with_CI(y_true, y_pred, y_lower, y_upper, index, fold):
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.scatter(index, y_true, label='Actual', color='black', marker='o')
    ax.plot(index, y_pred, label='Prediction (median)', color='green', linestyle='--', marker='o')
    ax.fill_between(index, y_lower, y_upper, color='green', alpha=0.1, label='80% CI')
    ax.set_ylim([0, 20000])
    ax.set_title(f"Fold {fold} - LightGBM Quantile Prediction with Confidence Interval - {str(index.min())[:10]} to {str(index.max())[:10]}")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()


# == using plotly ===
def plot_with_CI_plotly(y_true, y_pred, y_lower, y_upper, index, fold):
    # Convert everything to Series for safety
    df = pd.DataFrame({
        'ds': index,
        'y_true': y_true,
        'y_pred': y_pred,
        'y_lower': y_lower,
        'y_upper': y_upper
    })

    fig = go.Figure()

    # Actual values
    fig.add_trace(go.Scatter(
        x=df['ds'],
        y=df['y_true'],
        mode='markers',
        name='Actual',
        marker=dict(color='black', symbol='circle')
        #line=dict(color='black')
    ))

    # Predicted median
    fig.add_trace(go.Scatter(
        x=df['ds'],
        y=df['y_pred'],
        mode='lines',
        name='Prediction (median)',
        line=dict(color='blue')
        #marker=dict(symbol='circle', size=6)
    ))

    # Confidence interval: shaded region
    fig.add_trace(go.Scatter(
        x=pd.concat([df['ds'], df['ds'][::-1]]),
        y=pd.concat([df['y_upper'], df['y_lower'][::-1]]),
        fill='toself',
        fillcolor='rgba(70, 130, 180, 0.3)',  # steelblue with 30% opacity
        line=dict(color='rgba(255,255,255,0)'), # no border
        hoverinfo="skip",
        showlegend=True,
        name='80% CI'
    ))

    # Update layout
    fig.update_layout(
        title=f"Fold {fold} - LightGBM Quantile Prediction with Confidence Interval: from {str(index.min())[:10]} to {str(index.max())[:10]}",
        yaxis=dict(range=[0, 25000]),
        xaxis_title="Date",
        yaxis_title="Redemption Count",
        legend=dict(x=0.01, y=0.99),
        template="plotly_white",
        height=400,
        width=1600
    )

    fig.show()


# === 4. HYPERPARAMETER TUNING ===
def tune_lgbm_base_model_params(X, y, n_trials=25):
    """Tune XGBoost base model using Optuna and time-series CV."""
    y_log = np.log1p(y)
    tscv = TimeSeriesSplit(n_splits=3)

    def objective(trial):
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 100, 400, step=50),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, step=0.01),
            'max_depth': trial.suggest_int('max_depth', 4, 6),
            'subsample': trial.suggest_float('subsample', 0.7, 1.0),
            'colsample_bytree': trial.suggest_float('colsample_bytree', 0.7, 1.0),
            'reg_lambda': trial.suggest_int('reg_lambda', 0, 2),
            'reg_alpha': trial.suggest_int('reg_alpha', 0, 2),
            'min_data_in_leaf': 1,
            'min_child_samples': 1,
            'min_split_gain': 0.0
        }
        mae_scores = []
        for train_idx, val_idx in tscv.split(X):
            model = LGBMRegressor(**params, random_state=42)
            ## using log
            # model.fit(X.iloc[train_idx], y_log.iloc[train_idx])
            # preds = np.expm1(model.predict(X.iloc[val_idx]))
            
            model.fit(X.iloc[train_idx], y.iloc[train_idx])
            preds = model.predict(X.iloc[val_idx])
            mae_scores.append(mean_absolute_error(y[val_idx], preds))
        return np.mean(mae_scores)

    study = optuna.create_study(direction="minimize")
    study.optimize(objective, n_trials=n_trials)
    print("Best Base Model Params Found:")
    print(study.best_params)
    return study.best_params

# === Add Quantile Regressor Helper ===
def train_quantile_models(X_train, y_train, base_params, quantiles=[0.1, 0.5, 0.9]):
    models = {}
    for q in quantiles:
        model = LGBMRegressor(**base_params, objective='quantile', alpha=q, random_state=42)
        model.fit(X_train, y_train)
        models[q] = model
    return models


# === 5. MAIN FORECASTING PIPELINE ===
def run_lgbm_forecasting_pipeline(X, y, best_params, target_col="Sales Count", n_splits=3, test_size=7):
    tscv = TimeSeriesSplit(n_splits=n_splits, test_size=test_size)

    for fold, (train_idx, test_idx) in enumerate(tscv.split(X)):
        
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

        # y_log = np.log1p(y_train)
        
        # # # === Train Base Model using log values ===
        # base_model_log = LGBMRegressor(**best_params, random_state=42)
        # # Train with log value
        # base_model_log.fit(X_train, y_log)
        
        # # plot shap
        # # === base model shap values ===
        # # SHAP plot only on final fold or conditionally if desired
        # # explainer = shap.Explainer(base_model_log)
        # # shap_values0 = explainer0(X_train)

        # # shap.summary_plot(shap_values0, X_train, plot_type="bar")
        
        # # ==== Base prediction using log of y-train ====
        # base_pred_test_using_log = np.expm1(base_model_log.predict(X_test))
        # evaluate_model(y_test, base_pred_test_using_log, label="XGB Model log")
        # plot_predictions(y_test, base_pred_test_using_log, y_test.index, fold + 1)
        
        # === Train Quantile Models ===
        quantile_models = train_quantile_models(X_train, y_train, best_params, quantiles=[0.1, 0.5, 0.9])
        
        # # SHAP plot only on final fold or conditionally if desired
        # Get the median model (quantile 0.5)
        median_model = quantile_models[0.5]

        # Create SHAP TreeExplainer
        explainer = shap.Explainer(median_model)

        # Compute SHAP values
        shap_values = explainer(X_train)

        # Plot SHAP summary
        shap.summary_plot(shap_values, X_train)

        

        # Predict quantiles
        yhat_lower = quantile_models[0.1].predict(X_test)
        yhat_pred = quantile_models[0.5].predict(X_test)
        yhat_upper = quantile_models[0.90].predict(X_test)
        
        
        
        # Evaluate median model
        evaluate_model(y_test, yhat_pred, label="LGBM Quantile Model (Median)")

        # Plot with CI
        # plot_with_CI(y_test, yhat_pred, yhat_lower, yhat_upper, y_test.index, fold + 1)
        plot_with_CI_plotly(y_true=y_test, y_pred=yhat_pred, y_lower=yhat_lower, y_upper=yhat_upper, index=y_test.index, fold=fold+1)

        
# === 7. EXECUTE ===
if __name__ == "__main__":
    df = load_data('../data/Toronto Island Ferry Ticket Counts.csv')
   
    target_col = "Redemption Count"
    non_target_col = "Sales Count"
    
    plot_original_time_series(df, target_col)
    
    df = create_features(df, target_col, non_target_col="Sales Count")
    
    X = df.drop(columns=[target_col, non_target_col], errors='ignore')
    
    y = df[target_col]

    X = X.replace([np.inf, -np.inf], np.nan).dropna()
    y = y.loc[X.index]

    best_params = tune_lgbm_base_model_params(X, y, n_trials=15)
    
    forecast_horizon = 30 # in days
    run_lgbm_forecasting_pipeline(X, y, best_params=best_params, target_col=target_col, n_splits=3, test_size=forecast_horizon)
```
